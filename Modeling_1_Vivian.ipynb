```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
df_heat = pd.read_csv('heat_stats.csv')
df_heat = df_heat.dropna() # drops the NaN values but do we need them?
df_heat.head()
```

```python
df_lb = pd.read_csv('lebronstats.csv')
df_lb = df_lb.dropna()
df_lb.reset_index()
df_lb.head()
```

```python
df_lb[['PTS', 'AST', 'REB']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_lb[['PTS', 'AST', 'REB']], df_lb['WL'], test_size=0.4)
```

```python
rforestclf = RandomForestClassifier(n_estimators = 100)
# print(X_train.values)
# print(y_train.values.ravel())
```

```python
rforestclf.fit(X_train, y_train)
```

```python
results = rforestclf.predict(X_test)
print('SCORE: ', rforestclf.score(X_test, y_test)*100, '%')
```
