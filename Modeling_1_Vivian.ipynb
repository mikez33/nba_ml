```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df_heat = pd.read_csv('heat_stats.csv')
df_heat.head()
```

```python
df_lb = pd.read_csv('lebronstats.csv')
df_lb.reset_index()
df_lb.head()
```

```python
df_lb[['PTS', 'AST', 'REB']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_lb[['PTS', 'AST', 'REB']], df_lb['WL'], test_size=0.4)
```

```python
# X_train = pd.Series(X_train).to_numpy()
# X_test = pd.Series(X_test).to_numpy()
# y_train = pd.Series(y_train).to_numpy()
# y_test = pd.Series(y_test).to_numpy()
```

```python
rforestclf = RandomForestClassifier()
```

```python
rforestclf.fit(X_train.values, y_train.ravel())
```

```python
results = rforestclf.predict(X_test.values)
```

```python
total = 0
correct = 0
for i in range(len(results)):
    total += 1
    if results[i] == y_test.ravel()[i]:
        correct += 1
correct / total * 100
```
