```python
from nba_api.stats.endpoints import commonplayerinfo
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
import requests
```

```python
# Basic Request
player_info = commonplayerinfo.CommonPlayerInfo(player_id=2544)
```

```python
nba_teams = teams.get_teams()
# Select the dictionary for the Celtics, which contains their team ID
lakers = [team for team in nba_teams if team['abbreviation'] == 'LAL'][0]
lakers_id = lakers['id']
lakers
```

```python
# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=lakers_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
# Subset the games to when the last 4 digits of SEASON_ID were 2019.
games_1920 = games[games.SEASON_ID.str[-4:] == '2019']
games_1920.head()
```

```python
playergame = leaguegamefinder.LeagueGameFinder(player_id_nullable='2544')
playergame = playergame.get_data_frames()[0]
playergame
```

```python
playergame.to_csv('lebronstats.csv', index=False)
games_1920.to_csv('lakersstats.csv', index=False)
```
