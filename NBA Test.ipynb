```python
nba_teams = teams.get_teams()
# Select the dictionary for the Cavs, which contains their team ID
cavs = [team for team in nba_teams if team['abbreviation'] == 'CLE'][0]
cavs_id = cavs['id']
```

```python
# Query for games where the Cavs were playing
gamefinder_cavs = leaguegamefinder.LeagueGameFinder(team_id_nullable=cavs_id)
# The first DataFrame of those returned is what we want.
games_cavs = gamefinder_cavs.get_data_frames()[0]
games_cavs.head()
```

```python
cav_years = ['2006', '2007', '2008', '2009', '2010', '2015', '2016', '2017', '2018']
in_cavs = (games_cavs.SEASON_ID.str[-4:] == cav_years[0])|(games_cavs.SEASON_ID.str[-4:] == cav_years[1])|(games_cavs.SEASON_ID.str[-4:] == cav_years[2])|(games_cavs.SEASON_ID.str[-4:] == cav_years[3])|(games_cavs.SEASON_ID.str[-4:] == cav_years[4])|(games_cavs.SEASON_ID.str[-4:] == cav_years[5])|(games_cavs.SEASON_ID.str[-4:] == cav_years[6])|(games_cavs.SEASON_ID.str[-4:] == cav_years[7])|(games_cavs.SEASON_ID.str[-4:] == cav_years[8])
games_cavs_lebron = games_cavs[in_cavs]
games_cavs_lebron.head()
```

```python
games_cavs_lebron.to_csv('cavs_stats.csv', index=False)
```
